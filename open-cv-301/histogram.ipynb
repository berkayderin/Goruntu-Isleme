```python
import cv2 
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../img/cameraman.tif',0) # resimi okuduk ve 0 ile grayscale yaptık
histogram = cv2.calcHist([image], [0], None, [256], [0,256]) # image, channel, mask, histSize, range

plt.plot(histogram)
plt.show()
cv2.imshow('image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
